```python
import pandas as pd
poblacion= pd.read_csv('poblacion.csv', encoding='latin1', delimiter=';')
poblacion.head()
```

```python
#obtener la poblacion para 2024
poblacion_2024 = poblacion[poblacion["Año"]==2024]
poblacion_2024.head()
```

```python
#agrupar los datos por localidad
grupos= poblacion_2024.groupby("Localidad")
print(grupos)
```

```python
#sumar cada localidad
total_poblacion_localidad = grupos['Población']. sum()
print(total_poblacion_localidad)
```

```python
#se crea una serie a dataframe. El método reset_index() aplicado a la serie, cambia el indice por un indice numerico

df_poblacion_localidad = total_poblacion_localidad.reset_index()
df_poblacion_localidad
```

```python
#join dataframe
df_poblacion_localidad.rename(columns={"Localidad":"LocNombre"}, inplace=True)
df_poblacion_localidad
```

```python
#convertir la serie dataframe. El metodo es:
reset_index:df_poblacion_localidad= total_poblacion_localidad.reset_index()
df_poblacion_localidad.head()
```

```python
import pandas as pd

# Función para eliminar tildes
def replace_accents(input_str):

  replacements = {"Á": "A", "É": "E", "Í": "I", "Ó": "O", "Ú": "U"}
  for accented_char, unaccented_char in replacements.items():
    input_str = input_str.replace(accented_char, unaccented_char)
  return input_str

# Aplicar la función a la columna "LocNombre" del DataFrame
df_poblacion_localidad["LocNombreSinTilde"] = df_poblacion_localidad["LocNombre"].apply(replace_accents)

# Imprimir todas las localidades sin tilde
for localidad_sin_tilde in df_poblacion_localidad["LocNombreSinTilde"]:
  print(localidad_sin_tilde)

```
